```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r'../data/pima-indians.data.csv',low_memory=False, sep=',')
```

```python
df.dtypes
```

```python
df.head()
```

```python
import seaborn as sns
sns.set(style='whitegrid',context='notebook')
```

```python
cm = np.corrcoef(df.values.T)
sns.set(font_scale = 1.0)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10,10)
hm = sns.heatmap(cm,
                cbar = True,
                annot = True,
                square = True,
                fmt = '.2f',
                annot_kws = {'size': 12},
                yticklabels = df.columns,
                xticklabels = df.columns)
```
